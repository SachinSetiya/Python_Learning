```python
import numpy as nm
import matplotlib.pyplot as plt
import scipy as sc
```

```python
x1= nm.linspace(0,7,100)
x2= nm.linspace(7,15,100)
```

```python
y2= nm.random.randint(7,15,100)
```

```python
y_axis= nm.concatenate((nm.random.uniform(0,10,500), nm.random.uniform(0,10,500), nm.random.uniform(0,10,500)))
label_array= nm.concatenate((nm.full(500,1), nm.full(500,2), nm.full(500,3)))
x_axis= nm.concatenate((nm.linspace(0,4.9,500), nm.linspace(5.1, 9.9, 500), nm.linspace(10.1, 14.9, 500)))
new_array= [x_axis, y_axis, label_array]
```

```python
print(new_array[0])
print(new_array[1])
print(new_array[2])
```

```python
plt.figure()
count= 0
for a in new_array[2]:
  if a == 1:
    plt.plot(new_array[0][count], new_array[1][count], "r.")
  elif a == 2:
    plt.plot(new_array[0][count], new_array[1][count], "g.")
  else:
    plt.plot(new_array[0][count], new_array[1][count], "b.")
  count=count + 1
#plt.plot(new_array[0], new_array[1], "r.")
plt.show()
print(count)
```
